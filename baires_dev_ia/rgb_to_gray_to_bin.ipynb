Importando libs

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

Função para carregar imagem

```python
def load_image(path):
    img = Image.open(path)
    return np.array(img)
```

Função para converter rgb em niveis de cinza.

Esse valores estão ligados a percepção do olho humano (https://fr.mathworks.com/help/matlab/ref/rgb2gray.html)

```python
def from_rgb_to_gray(image):
    gray = 0.299 * image[:,:,0] + 0.587 * image[:,:,1] + 0.114 * image[:,:,2]
    return gray.astype(np.uint8)
```

Função para converter niveis de cinza em binario. Dividindo 256 / 2 = 128 usado como divisor entre preto e branco

```python
def from_gray_to_bin(gray_image):
    threshold = 128
    image = (gray_image > threshold) * 255
    return image.astype(np.uint8)
```

Função para mostrar as 3 imagens

```python
def show_images(rgb_image, gray_image, baw_image):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    axes[0].imshow(rgb_image)
    axes[0].set_title("rgb image")
    axes[0].axis('off')

    axes[1].imshow(gray_image, cmap='gray')
    axes[1].set_title("gray image")
    axes[1].axis('off')

    axes[2].imshow(baw_image, cmap='gray')
    axes[2].set_title("binary image")
    axes[2].axis('off')

    plt.show()
```

Codigo "main". Carrega o caminho da imagem, carrega a imagem, depois faz a conversão e 'plota' elas

```python
image_path = '/content/Lenna_(test_image).png'
image_path2 = '/content/morumbis.jpeg'
```

```python
rgb_image = load_image(image_path)

gray_image = from_rgb_to_gray(rgb_image)

baw_image = from_gray_to_bin(gray_image)

show_images(rgb_image, gray_image, baw_image)

#

rgb_image2 = load_image(image_path2)

gray_image2 = from_rgb_to_gray(rgb_image2)

baw_image2 = from_gray_to_bin(gray_image2)

show_images(rgb_image2, gray_image2, baw_image2)
```
